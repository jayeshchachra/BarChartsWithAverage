```python
import matplotlib.pyplot as plt
import numpy as np

names = np.array(["John", "Martin", "Chris", "Marshal"])
bst_marks = np.array([80,45,67,89])

plt.subplot(2, 3, 1)
plt.bar(names,bst_marks, width = 0.3, color = 'purple')
plt.title("Business Studies")
plt.ylabel("Student Marks")
plt.xlabel("Student Names")
plt.tight_layout()
plt.xticks(rotation=45)
bst_avg = np.average(bst_marks)


acc_marks = np.array([70, 56, 62, 36])

plt.subplot(2, 3, 2)
plt.bar(names,acc_marks, width = 0.3,color = 'green')
plt.title("Accountancy")
plt.ylabel("Student Marks")
plt.xlabel("Student Names")
plt.tight_layout()
plt.xticks(rotation=45)
acc_avg = np.average(acc_marks)


eco_marks = np.array([54, 59, 87, 40])

plt.subplot(2, 3, 3)
plt.bar(names,eco_marks, width = 0.3, color = 'red')
plt.title("Economics")
plt.ylabel("Student Marks")
plt.xlabel("Student Names")
plt.tight_layout()
plt.xticks(rotation=45)
eco_avg = np.average(eco_marks)

eng_marks = np.array([70, 90, 60, 47])

plt.subplot(2, 3, 5)
plt.bar(names,eng_marks, width = 0.3, color = 'magenta')
plt.title("English")
plt.ylabel("Student Marks")
plt.xlabel("Student Names")
plt.tight_layout()
plt.xticks(rotation=45)
eng_avg = np.average(eng_marks)

plt.subplot(2,3,6)
labels = ["Business Studies", "Accountancy", "Economics", "English"]
colors = ["purple", "green", "red", "magenta"]
plt.pie([bst_avg, acc_avg, eco_avg, eng_avg], labels = labels, colors = colors)
plt.legend(fontsize = 6)




plt.show() 
```
